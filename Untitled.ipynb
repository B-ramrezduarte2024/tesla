```python
# Importar las bibliotecas necesarias
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la página con datos de ingresos de GameStop
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Configurar headers para evitar bloqueos
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
response = requests.get(url, headers=headers)

# Verificar si se accedió correctamente a la página
if response.status_code != 200:
    print("Error al acceder a la página:", response.status_code)
else:
    soup = BeautifulSoup(response.content, "html.parser")

    # Buscar todas las tablas en la página
    tables = soup.find_all("table")
    print(f"Se encontraron {len(tables)} tablas en la página.")

    # Intentar localizar la tabla de ingresos
    revenue_table = None
    for table in tables:
        if "GameStop Quarterly Revenue" in str(table):
            revenue_table = table
            break

    if revenue_table is None:
        print("No se encontró la tabla de ingresos.")
    else:
        # Extraer los datos de la tabla
        rows = revenue_table.find_all("tr")
        data = []
        for row in rows[1:]:  # Saltar el encabezado
            cols = row.find_all("td")
            data.append([col.text.strip() for col in cols])

        # Crear un DataFrame
        columns = ["Date", "Revenue"]
        gme_revenue = pd.DataFrame(data, columns=columns)

        # Limpiar los datos (remover comas y caracteres innecesarios)
        gme_revenue["Revenue"] = gme_revenue["Revenue"].str.replace(",", "").str.replace("$", "")
        gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")

        # Mostrar las últimas cinco filas
        print(gme_revenue.tail())
```

```python
# Importar las bibliotecas necesarias
import matplotlib.pyplot as plt
import pandas as pd
import yfinance as yf

# Descargar los datos de GameStop
gme_data = yf.download("GME", start="2022-01-01", end="2023-01-01")

# Restablecer el índice
gme_data.reset_index(inplace=True)

# Verificar las columnas disponibles
print("Columnas disponibles:", gme_data.columns)

# Confirmar si 'Close' existe
if 'Close' not in gme_data.columns:
    print("Error: La columna 'Close' no está disponible.")
else:
    # Función para graficar
    def make_graph(data, title):
        plt.figure(figsize=(12, 6))
        plt.plot(data['Date'], data['Close'], label='GameStop Stock Price', color='red')
        plt.title(title)
        plt.xlabel('Date')
        plt.ylabel('Closing Price (USD)')
        plt.grid(True)
        plt.legend()
        plt.show()  # Forzar la visualización del gráfico

    # Crear el gráfico
    make_graph(gme_data, "GameStop Stock Price (2022-2023)")
```
